```python
sns.set_style('darkgrid')
```

```python
sns.histplot(penguins.bill_length_mm) #Why is this culmen_length_mm in the videos?!?!?!
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins, kde=True)
```

```python
#Alternate syntax
sns.histplot(y = 'bill_length_mm', data=penguins, kde=False)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins, bins=[30, 40, 45, 50, 60])
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    binwidth=5, binrange=(30,60)
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    stat='count'
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    stat='density' #area sum to 1
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    stat='probability', fill=False, element='step', 
    cumulative=True #probability height sum to 1
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    hue='species' #Split up data by columns with hue (other colors)
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    hue='species', #Split up data by columns with hue (other colors)
    element='step'
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    hue='species', #Split up data by columns with hue (other colors)
    element='poly'
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    hue='species', #Split up data by columns with hue (other colors)
    multiple='stack' # default'layer'
)
```

```python
#Alternate syntax
sns.histplot(x = 'bill_length_mm', data=penguins,
    hue='species', #Split up data by columns with hue (other colors)
    multiple='fill' # default'layer' #proportion
)
```

Bivariate Histplot

```python
#Two variables, no more scale so add color bar
sns.histplot(x = 'bill_length_mm', y='bill_depth_mm', data=penguins,
    cbar=True
)
```

```python
#Two variables, no more scale so add color bar ADD HUE
sns.histplot(x = 'bill_length_mm', y='bill_depth_mm', data=penguins,
   hue='species' #overlap is bad and hard to see. 
)
```

```python
#Two variables, no more scale so add color bar ADD HUE. Compare distributions of species in a visual way.
sns.histplot(x = 'bill_length_mm', y='species', data=penguins,
   hue='species' #overlap is bad and hard to see. 
)
```
