Median and Inner Quartile Range is the same. Kernel Density Estimation. Box plot shows outliers instead.  Extended distribution with whiskers too. 

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
#basics
sns.set_style('whitegrid')
cars.cylinders.value_counts()
```

```python
#2:00
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
#Alternative way
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars) #different car codes for origins of cars (new colors)
```

```python
#Symmetric but doesn't convey additional info until...
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars[cars.origin.isin(['japan', 'europe'])]
) 
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
    data=cars[cars.origin.isin(['japan', 'europe'])],
    split=True #only two (Japan and Europe)
)

plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
    data=cars[cars.origin.isin(['japan', 'europe'])],
    split=True, #only two (Japan and Europe)
    inner='quartiles'
)

plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
    data=cars[cars.origin.isin(['japan', 'europe'])],
    split=True, #only two (Japan and Europe)
    scale='count' #see observations for each portion of your data
    #split then scale
)

plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
    data=cars[cars.origin.isin(['japan', 'europe'])],
    split=True, #only two (Japan and Europe)
    scale='count', #see observations for each portion of your data
    scale_hue=False
)

plt.legend(loc=2)
```

```python
( #get the number of observations
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

```python
##FINAL SEGMENT
#styling options
sns.violinplot(x=cars.origin, y=cars.displacement,
    order=['japan', 'europe', 'usa']
)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement, 
    linewidth=3
    )
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement, 
    bw=0.2
)
```
